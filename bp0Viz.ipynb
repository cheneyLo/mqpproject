```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
images_per_filter = 9
filters = 16
plt.figure(figsize=(10, 10))
for filter_number in range(filters):
    for image_number in range(images_per_filter):
        image = bp0.X_train[bp0.X_max[filter_number][image_number]]
        plt.subplot(16,9,filter_number*images_per_filter+image_number+1)
        plt.ylabel("{:}".format(filter_number))
        plt.gca().get_xaxis().set_ticks([])
        plt.gca().get_yaxis().set_ticks([])
        plt.imshow(image[0], cmap='gray')
```
